# Пуассоновский процесс

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def poisson_process(mu: float, num_events: int) -> pd.DataFrame:
    time_intervals = -np.log(np.random.random(num_events)) / mu
    total_events = time_intervals.cumsum()
    events = pd.DataFrame(np.ones(num_events), index=total_events)
    events[0] = events[0].cumsum()
    return events
```

```python
gamma = 15
num_events_simulated = 50

proc = poisson_process(gamma, num_events_simulated)
plt.step(proc[0].values, proc[0].index)
plt.grid(True)
plt.title('Траектория пуассоновского процесса')
plt.xlabel('Время')
plt.ylabel('События')
plt.show()
```
